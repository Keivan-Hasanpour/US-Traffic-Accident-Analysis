### Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

city_encoder = LabelEncoder()

df["City"] = city_encoder.fit_transform(df["City"].to_numpy())
df["City"] = df["City"].astype(np.uint16)
```

```python
df = df.drop(columns=["ID", "Source", "Country", "Description"])
```

```python
df["Start_Time"] = pd.to_datetime(df["Start_Time"].str[:19])
df["End_Time"] = pd.to_datetime(df["End_Time"].str[:19])
```

```python
df["Timezone"] = df["Timezone"].map(
    {
        "US/Eastern": 0,
        "US/Pacific": 1,
        "US/Central": 2,
        "US/Mountain": 3
    }
)
```

```python
df.info()
```

```python
df["Severity"] = df["Severity"].astype(np.uint8)
```

```python
df["Astronomical_Twilight"].value_counts()
```

```python
for col in ["Sunrise_Sunset", "Civil_Twilight", "Nautical_Twilight", "Astronomical_Twilight"]:
  df[col] = (
    df[col]
    .map({"Night": 0, "Day": 1})
    .fillna(-1)
    .astype(np.int8)
)
```

```python
df.info()
```

```python
df["Wind_Direction"].value_counts()
```

```python
wind_encoder = LabelEncoder()

df["Wind_Direction"] = city_encoder.fit_transform(df["Wind_Direction"].to_numpy())
df["Wind_Direction"] = df["Wind_Direction"].astype(np.uint16)
```

```python

for col in ['Distance(mi)',
       'Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)',
       'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)']:
  df[col] = df[col].astype(np.float16)
```

```python
df.info()
```

```python
df.to_csv("us_accidents.csv")
```

### Data Visulization

#### Number of accidents in different states

```python
state_counts = df["State"].value_counts()

fig = px.bar(
    x=state_counts.index,
    y=state_counts.values,
    labels={"x": "State", "y": "Number of accidents"},
    title="Number of accidents in different states"
)

fig.update_layout(xaxis_tickangle=-90)

fig.show()
```

#### Top 10 States with the Highest Number of Accidents

```python
state_counts = df["State"].value_counts().head(10)

fig = px.bar(
    x=state_counts.values,
    y=state_counts.index,
    orientation="h",
    color=state_counts.values,
    color_continuous_scale="Viridis",
    labels={"x": "Number of Accidents", "y": "State"},
    title="Top 10 States with the Highest Number of Accidents"
)

fig.update_layout(
    yaxis=dict(categoryorder="total ascending")
)

fig.show()
```

#### 5 Top Weather Conditions for Accidents

```python
weather_counts = df['Weather_Condition'].value_counts().head(5)

fig = px.bar(
    x=weather_counts.index,
    y=weather_counts.values,
    labels={"x": "Weather Condition", "y": "Number of Accidents"},
    title="5 Top Weather Conditions for Accidents"
)

fig.update_layout(
    xaxis_tickangle=-90
)

fig.show()
```

#### Count of Severity

```python
severity_counts = df['Severity'].value_counts()

fig = make_subplots(
    rows=1, cols=2,
    specs=[[{"type": "domain"}, {"type": "xy"}]],
    subplot_titles=("Percentage Severity Distribution", "Count of Severity")
)

fig.add_trace(
    go.Pie(
        labels=severity_counts.index,
        values=severity_counts.values,
        hole=0,
    ),
    row=1, col=1
)


fig.add_trace(
    go.Bar(
        x=severity_counts.index,
        y=severity_counts.values
    ),
    row=1, col=2
)

fig.update_layout(
    title_text="Severity Analysis",
    showlegend=False
)

fig.show()
```

####  Medium distance by severity

```python
severity_distance = (
    df.groupby("Severity")["Distance(mi)"]
    .mean()
    .sort_values(ascending=False)
)

fig = px.bar(
    x=severity_distance.values,
    y=severity_distance.index,
    orientation="h",
    color=severity_distance.values,
    color_continuous_scale="Viridis",
    labels={
        "x": "Distance (mi)",
        "y": "Severity"
    },
    title="Mean distance by severity"
)

fig.update_layout(
    yaxis=dict(categoryorder="total ascending")
)

fig.show()
```

####  Number of accidents for weekday


```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

counts = df['Start_Time'].dt.day_name().value_counts()

weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

counts = counts.reindex(weekdays)

fig = px.bar(
    x=counts.index,
    y=counts.values,
    color=counts.values,
    color_continuous_scale="Viridis",
    labels={
        "x": "Weekday",
        "y": "Number of Accidents"
    },
    title="Number of accidents for each weekday"
)

fig.update_layout(
    xaxis_tickangle=-45
)

fig.show()
```

```python
df_num = df[['Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 'Severity',
             'Start_Lng', 'End_Lng', 'Start_Lat', 'End_Lat',
             'Distance(mi)', 'Pressure(in)', 'Visibility(mi)',
             'Wind_Speed(mph)', 'Precipitation(in)']]

corr = df_num.corr()

fig = px.imshow(
    corr,
    text_auto=".2f",
    color_continuous_scale="RdBu_r",
    zmin=-1,
    zmax=1,
    aspect="auto",
    title="Correlation Heatmap"
)

fig.update_layout(
    width=900,
    height=800
)

fig.show()
```

####  Severity by Fog, Light Rain, Rain, Heavy Rain and Snow

```python
weather_conditions = ["Fog", "Light Rain", "Rain", "Heavy Rain", "Snow"]

for s in weather_conditions:

    filtered = df[df["Weather_Condition"] == s]["Severity"].value_counts()

    fig = make_subplots(
        rows=1, cols=2,
        specs=[[{"type": "xy"}, {"type": "domain"}]],
        subplot_titles=("Severity Count", "Severity Percentage")
    )


    fig.add_trace(
        go.Bar(
            x=filtered.index,
            y=filtered.values,
            marker_color="gold",
            name="Count"
        ),
        row=1, col=1
    )


    fig.add_trace(
        go.Pie(
            labels=filtered.index,
            values=filtered.values,
            textinfo="percent+label"
        ),
        row=1, col=2
    )

    fig.update_layout(
        title_text=f"Accident Severity Under {s}",
        showlegend=False
    )

    fig.show()
```

```python
features = [
    'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit',
    'Railway', 'Roundabout', 'Station', 'Stop',
    'Traffic_Calming', 'Traffic_Signal', 'Turning_Loop'
]

for s in features:


    if s in df.columns and df[s].sum() > 0:

        severity_counts = df[df[s] == True]['Severity'].value_counts()

        fig = make_subplots(
            rows=1, cols=2,
            specs=[[{"type": "xy"}, {"type": "domain"}]],
            subplot_titles=("Severity Count", "Severity Percentage")
        )


        fig.add_trace(
            go.Bar(
                x=severity_counts.index,
                y=severity_counts.values,
                marker_color="gold"
            ),
            row=1, col=1
        )


        fig.add_trace(
            go.Pie(
                labels=severity_counts.index,
                values=severity_counts.values,
                textinfo="percent+label"
            ),
            row=1, col=2
        )

        fig.update_layout(
            title=f"Accident Severity Near {s}",
            showlegend=False
        )

        fig.show()
```

#### Accident Location Distribution

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "notebook_connected"

df_clean = df.dropna(subset=["Start_Lat", "Start_Lng"]).sample(1000000, random_state=42)

df_clean["Start_Lat"] = df_clean["Start_Lat"].round(2)
df_clean["Start_Lng"] = df_clean["Start_Lng"].round(2)

fig = px.scatter_mapbox(
    df_clean,
    lat="Start_Lat",
    lon="Start_Lng",
    color="Severity",
    opacity=0.3,
    zoom=4,
    mapbox_style="open-street-map",
    title="Accident Locations (Optimized)"
)

fig.show()
```
